# Business Understanding

This data set contains records about over 16,000 video games released in 2016 or earlier. The data includes attributes with basic facts about the game (year of release, genre, etc), sales data in different parts of the world, and review data. The data was collected via a web scraper from two different sources: VGChartz for the basic facts and sales data, and Metacritic for the review data. This data can be used to predict how many copies of a game will be sold based on the type of game and its reviews. The problem with this data set is that the review data is lagged, meaning that it is released after or shortly before a game is launched, meaning that predictions from it are not helpful to the developer. However, video games go through a rigorous testing process, in which external testers are brought in during the beta testing stage (1). These beta testers generally represent the intended target audience, but can also include content creators and reviewers. Through the testing process, studios can get feedback and early reviews for their games while they are in early development. Such a data set is not publically available, but it would most likely be similar in structure to the data used in this notebook. This data set will be used to simulate early feedback reviews that a studio would generate through beta testers. If a game receives negative enough feedback from beta testers, the development of the game may be cancelled so that the studio can move on to other projects, like what happened to *Star Wars: Attack Squadrons* in 2014 (2). A prediction of how many sales a game will make based on early reviews and the type of game can help a studio determine if it is worth continuing production of a game. To be a useful prediction, a model would need to be able to, more often that not, accurately predict whether or not the game will generate enough sales to make finishing the development process worthwhile.

# Load Data

```python
# Load raw data from csv file
df = pd.read_csv("../Datasets/video_game_sales.csv")

df.head()
```

For prediction purposes, only global sales are needed. Thus, sales numbers from specific regions can be dropped. Additionally, the number of critics or users who reviewed a game on Metacritic is not useful if this data set is meant to be an estimate of early reviews, so those attributes can also be dropped. Finally, the name of the game only serves as a label, so it is not needed for prediction and can be dropped.

```python
# Drop unwanted columns
df = df.drop(["Name", "NA_Sales", "EU_Sales", "JP_Sales", "Other_Sales", "Critic_Count", "User_Count"], axis = 1)

df.info()
```

# Clean Data

Based on the data types assigned to each column, the only one that does not seem right is the user score. The user score should be numeric, not an object, so let's see what non-numeric values are in this column.

```python
# Find any non-numeric values in the user score column (max possible user score is 9.9)
df[df["User_Score"] > "9.9"]["User_Score"].value_counts()
```

2,425 records have a user score of "tbd", which is not useful for analysis, so these records can be dropped.

```python
# Drop tbd values from the user score column
df = df[df["User_Score"] != "tbd"]
```

```python
%matplotlib inline

# Visualize missing data
mn.matrix(df.sort_values(["Critic_Score", "User_Score", "Developer", "Rating", "Year_of_Release", "Publisher"]))

# Print the number of records with and without missing data
print("Number of records with missing values:", len(df) - len(df.dropna()))
print("Number of records after dropping missing values:", len(df.dropna()))
```

Based on the plot above, there are a significant number of records missing critic score, user score, developer, and rating data. This missing values exist, because the data for these games was found on VGChartz but not Metacritic. There are also some other records that are missing data in some columns. If all records with any missing data are dropped, there are still 6,825 records remaining, which should be enough data to detect any patterns useful for prediction.

After dropping the missing data, the data type user score column can be changed to float and the data type of the year of release and critic score columns can be changed to int.

```python
# Drop records with missing values
df = df.dropna().reset_index(drop = True)
# Convert the user score to a numeric column after removing 'tbd' and missing values
df["User_Score"] = pd.to_numeric(df["User_Score"])
# Convert year of release column to integer (ordinal feature)
df["Year_of_Release"] = df["Year_of_Release"].astype(int)
df.head()
```

### Are critic or user scores more correlated with global sales?

```python
# Remove all non-numeric columns
df_numeric_feature = df.select_dtypes(include='number')

f, ax = plt.subplots(figsize=(8, 8))
sns.heatmap(df_numeric_feature.corr(), annot=True)
 
f.tight_layout()
```

The correlation plot above shows the Pearson's correlation between each pair of quantitative features in this data set. The plot shows a few things about the correlation between variables:

- The critic score of a game is very positively correlated with the user score of that game (r = 0.58)
- The critic score of a game is more correlated with the game's global sales than the user score (r = 0.23 vs r = 0.081)
- The user score of a game is negatively correlated with the year of release (r = -0.24), suggesting that more recent games get worse use scores than older games
- There is not a significant correlation between a game's critic score and the year of release (r = 0.016)

The problem with Pearson's correlation is that it can have trouble detecting nonlinear relationships. Observing plots of the critic and user scores can help determine if this is an issue here.

```python
sns.lineplot(x = "Critic_Score", y = "Global_Sales", data = df)

plt.title("Mean Global Video Game Sales By Critic Score")
plt.xlabel("Critic Score")
plt.ylabel("Global Video Game Sales (Millions of Copies)")

plt.show()
```

The line plot above shows the mean global sales and 95% confidence interval plotted against the critic score. The plot shows that the relationship between the mean global sales and critic score is exponential, not linear, suggesting that there may be a stronger correlation between these variables than the Pearson's correlation indicates.

```python
sns.lineplot(x = "User_Score", y = "Global_Sales", data = df)

plt.title("Mean Global Video Game Sales By User Score")
plt.xlabel("User Score")
plt.ylabel("Global Video Game Sales (Millions of Copies)")

plt.show()
```

The line plot above shows the mean global sales and 95% confidence interval plotted against the user score.
```python
f, ax = plt.subplots(figsize=(8, 8))
sns.heatmap(df_numeric_feature.corr(method = "spearman"), annot=True)
 
f.tight_layout()
```

The correlation plot above shows the Spearman's correlation between each pair of quantitative variables in this data set. Spearman's correlation is better at detecting nonlinear correlations than Pearson's correlation, because Spearman's attempts to find a linear relationship between the ranks of the observations instead of the raw values. This means that the correlation increases if both variables consistently increase/decrease together in any way, linear or not. 

The correlation plot shows that the Spearman's correlation between the critic score and the global sales is greater than the Pearson's correlation (r = 0.38 vs r = 0.23). The Spearman's correlation between the user score and the global sales is also greater than the Pearson's correlation (r = 0.15 vs 0.081), but is lower than the correlation between the critic score and global sales.

## UMAP

Uniform Manifold Approximation and Projection (UMAP) is a dimensionality reduction algorithm that makes it easier to visualize data sets with a large number of features. To do this, UMAP constructs a graph of the data set with the full dimensionality, and then optimizes a graph with a lower dimensionality to have as similar a structure to the original graph as possible. Although algorithms like t-SNE existed before UMAP, these algorithms were much slower than UMAP and were generally worse at preserving the global structure of the projection. UMAP can be used to visualize large dimensionality data sets, and attempt to find clusters within the data by color coding a target variable.

Sources:
- https://pair-code.github.io/understanding-umap/

```python
# Create UMAP dimensionality reduction object
reducer = umap.UMAP()

# Scale numeric features
scaled_data = StandardScaler().fit_transform(df_numeric_feature)
# Create UMAP embedding of numeric features
embedding = reducer.fit_transform(scaled_data)
```

```python
# Plot UMAP embedding
plt.scatter(
    embedding[:,0], 
    embedding[:,1],
    color = [
        # Color code genre values
        sns.color_palette()[x % len(sns.color_palette())] for x in df["Genre"].map({
            y: i for i, y in enumerate(set(df["Genre"]))
        })
    ]
)
plt.gca().set_aspect('equal', 'datalim')
plt.title("UMAP Projection of Video Game Sales By Genre")

plt.show()
```

```python
# Plot UMAP embedding
plt.scatter(
    embedding[:,0], 
    embedding[:,1],
    color = [
        # Color code rating values
        sns.color_palette()[x % len(sns.color_palette())] for x in df["Rating"].map({
            y: i for i, y in enumerate(set(df["Rating"]))
        })
    ]
)
plt.gca().set_aspect('equal', 'datalim')
plt.title("UMAP Projection of Video Game Sales By Rating")

plt.show()
```

```python
plt.scatter(
    embedding[:,0], 
    embedding[:,1],
    color = [
        # Color code platform values
        sns.color_palette()[x % len(sns.color_palette())] for x in df["Platform"].map({
            y: i for i, y in enumerate(set(df["Platform"]))
        })
    ]
)
plt.gca().set_aspect('equal', 'datalim')
plt.title("UMAP Projection of Video Game Sales By Platform")

plt.show()
```

Attempting to color code a UMAP representation of this data set by genre, rating, and platform in the above plots shows that there are no significant clusters by these features.
